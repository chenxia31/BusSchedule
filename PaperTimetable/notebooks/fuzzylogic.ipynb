```python
import numpy as np
import matplotlib.pyplot as plt

# Define the membership functions
x = np.linspace(0, 10, 1000)

# Trapezoidal membership function
def trapmf(x, a, b, c, d):
    return np.maximum(np.minimum((x - a) / (b - a), (d - x) / (d - c)), 0)

# Square (or rectangular) membership function
def squaremf(x, a, b):
    return np.where((a <= x) & (x <= b), 1, 0)

# Triangular membership function
def trimf(x, a, b, c):
    return np.maximum(np.minimum((x - a) / (b - a), (c - x) / (c - b)), 0)

# Define the parameters for the membership functions
trapmf_params = [2, 4, 6, 8]
squaremf_params = [3, 7]
trimf_params = [2, 5,8]

# Calculate the membership values
trapmf_values = trapmf(x, *trapmf_params)
squaremf_values = squaremf(x, *squaremf_params)
trimf_values = trimf(x, *trimf_params)

# Correct the trapezoidal membership function
def trapmf_corrected(x, a, b, c, d):
    return np.where(x < a, 0, np.where(x < b, (x - a) / (b - a), np.where(x < c, 1, np.where(x <= d, (d - x) / (d - c), 0))))

# Calculate the new membership values
trapmf_values_corrected = trapmf_corrected(x, *trapmf_params)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot the corrected trapezoidal membership function
axs[0].plot(x, trapmf_values_corrected, label='Trapezoidal MF', color='blue')
axs[0].set_title('Trapezoidal Membership Function')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Membership Degree')
axs[0].legend()
axs[0].grid(True)

# Plot the square membership function
axs[1].plot(x, squaremf_values, label='Square MF', color='green')
axs[1].set_title('Square Membership Function')
axs[1].set_xlabel('x')
axs[1].set_ylabel('Membership Degree')
axs[1].legend()
axs[1].grid(True)

# Plot the triangular membership function
axs[2].plot(x, trimf_values, label='Triangular MF', color='red')
axs[2].set_title('Triangular Membership Function')
axs[2].set_xlabel('x')
axs[2].set_ylabel('Membership Degree')
axs[2].legend()
axs[2].grid(True)

# Show the plot
plt.savefig('../../output/membership.pdf',bbox_inches='tight')
```
